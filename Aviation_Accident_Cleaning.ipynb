#### Data Cleaning – Filtering Aircraft and Events

To meet the client’s needs, we filter the dataset to include only events relevant to commercial and passenger aircraft with potential modern relevance.

- Only include events from 1983 onward
- Filter for aircraft with category "Airplane"
- Filter for professionally built aircraft (Amateur.Built = "No")

```python
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
```

```python
# Filter 1: Remove events before 1983
df = df[df['Event.Date'] >= '1983-01-01']

# Filter 2: Keep only airplanes
df = df[df['Aircraft.Category'] == 'Airplane']

# Filter 3: Keep only professional builds (Amateur.Built == 'No')
df = df[df['Amateur.Built'] == 'No']

df.head()
```

### Constructing Fatal/Serious Injury Metric

We estimate the number of passengers per event by summing all injury and uninjured counts.  
We then calculate the injury rate as the fraction of passengers who were either seriously or fatally injured.

NaNs in the injury columns are treated as 0s. Any rows where the total number of passengers is 0 are excluded from the rate calculation.

```python
# First, fill NaNs with 0s for injury-related columns
injury_cols = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 
               'Total.Minor.Injuries', 'Total.Uninjured']
df[injury_cols] = df[injury_cols].fillna(0)

# Estimate total people onboard
df['Estimated.Passengers'] = df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries'] + \
                             df['Total.Minor.Injuries'] + df['Total.Uninjured']

# Avoid divide-by-zero
df['Estimated.Passengers'] = df['Estimated.Passengers'].replace(0, np.nan)

# Construct injury rate metric
df['Injury.Rate'] = (df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries']) / df['Estimated.Passengers']
df.head()
```

Cleaning the Make Column

We inspect and clean the `Make` column to ensure consistency in manufacturer names.


```python
# Inspect first
df['Make'].value_counts(dropna=False).head(20)

# Step 1: Clean formatting
df['Make'] = df['Make'].str.upper().str.strip()

# Step 2: Count occurrences
make_counts = df['Make'].value_counts()

# Step 3: Filter to Makes with >= 50 records
valid_makes = make_counts[make_counts >= 50].index
df = df[df['Make'].isin(valid_makes)]
df.head()
```

Inspecting and Cleaning the Model Column

We remove any missing `Model` values and standardize the formatting. Upon inspection, some model names are shared across different manufacturers, so we create a new column `Plane.Type` by combining the `Make` and `Model` columns to uniquely identify each aircraft type.


```python
# Drop rows where Model is NaN
df = df[df['Model'].notna()]

# Optional: Clean the Model column formatting
df['Model'] = df['Model'].str.upper().str.strip()

# Preview top model values
df['Model'].value_counts().head(10)
```

```python
# Count how many unique makes exist per model
model_make_check = df.groupby('Model')['Make'].nunique().sort_values(ascending=False)

# Show top models used by multiple makes
model_make_check[model_make_check > 1].head(10)
```

```python
# Combine Make and Model into one column
df['Plane.Type'] = df['Make'] + ' ' + df['Model']

# Show a few rows to confirm
df[['Make', 'Model', 'Plane.Type']].head()
```
